```python
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
import hashlib
import os

# Função de criptografia com Kyber
def kyber_encrypt_decrypt():
    # Gerar a chave compartilhada (shared_secret)
    shared_secret = os.urandom(32)  # Exemplo de chave simulada (você deve obter da função Kyber)

    # Texto que será criptografado
    plaintext = b"Teste de Criptografia"

    # Gerar um nonce aleatório
    nonce = get_random_bytes(12)  # Nonce de 12 bytes para AES GCM

    # Gerar o cipher AES com GCM para criptografia
    cipher = AES.new(shared_secret, AES.MODE_GCM, nonce=nonce)

    # Criptografar o texto
    ciphertext, tag = cipher.encrypt_and_digest(plaintext)

    print("===== Criptografia =====")
    print(f"Shared Secret: {shared_secret.hex()}")
    print(f"Nonce: {nonce.hex()}")
    print(f"Ciphertext: {ciphertext.hex()}")
    print(f"Tag: {tag.hex()}")

    # Descriptografar usando a mesma chave, nonce e verificar o tag
    cipher_dec = AES.new(shared_secret, AES.MODE_GCM, nonce=nonce)
    decrypted_text = cipher_dec.decrypt_and_verify(ciphertext, tag)

    print("\n===== Descriptografia =====")
    print(f"Decrypted Text: {decrypted_text.decode('utf-8')}")

# Executar a função
kyber_encrypt_decrypt()
```

```python
import time
import psutil
import os
from Crypto.Cipher import AES
from Crypto.Random import get_random_bytes
from binascii import unhexlify

# Função para medir tempo, CPU e memória
def measure_performance(func, *args, **kwargs):
    start_time = time.time()

    process = psutil.Process(os.getpid())
    start_memory = process.memory_info().rss / 1024 / 1024  # Convertendo para MB
    start_cpu = process.cpu_percent(interval=None)  # CPU inicial

    result = func(*args, **kwargs)

    end_time = time.time()

    end_memory = process.memory_info().rss / 1024 / 1024  # Convertendo para MB
    end_cpu = process.cpu_percent(interval=None)  # CPU final

    elapsed_time = end_time - start_time
    memory_used = end_memory - start_memory
    cpu_used = end_cpu - start_cpu

    print(f"Tempo de execução: {elapsed_time:.4f} segundos")
    print(f"Memória utilizada: {memory_used:.4f} MB")
    print(f"CPU utilizada: {cpu_used:.2f}%\n")

    return result

# Função para criptografar um texto
def encrypt_text():
    shared_secret = os.urandom(32)  # Simulação da chave gerada (deve ser obtida pelo Kyber)

    # Entrada do texto pelo usuário
    plaintext = input("Digite o texto que deseja criptografar: ").encode('utf-8')

    nonce = get_random_bytes(12)  # Nonce de 12 bytes

    cipher = AES.new(shared_secret, AES.MODE_GCM, nonce=nonce)
    ciphertext, tag = cipher.encrypt_and_digest(plaintext)

    # Mostrando as informações necessárias para a descriptografia
    print("===== Criptografia Completa =====")
    print(f"Shared Secret (hex): {shared_secret.hex()}")
    print(f"Nonce (hex): {nonce.hex()}")
    print(f"Ciphertext (hex): {ciphertext.hex()}")
    print(f"Tag (hex): {tag.hex()}")

    return shared_secret, nonce, ciphertext, tag

# Função para descriptografar o texto
def decrypt_text():
    # Solicitar ao usuário as informações necessárias
    shared_secret_hex = input("Digite a chave compartilhada (Shared Secret) em hex: ")
    nonce_hex = input("Digite o Nonce em hex: ")
    ciphertext_hex = input("Digite o Ciphertext em hex: ")
    tag_hex = input("Digite o Tag em hex: ")

    # Converter os valores hexadecimais de volta para bytes
    shared_secret = unhexlify(shared_secret_hex)
    nonce = unhexlify(nonce_hex)
    ciphertext = unhexlify(ciphertext_hex)
    tag = unhexlify(tag_hex)

    # Descriptografar usando as informações fornecidas
    try:
        cipher = AES.new(shared_secret, AES.MODE_GCM, nonce=nonce)
        decrypted_text = cipher.decrypt_and_verify(ciphertext, tag)
        print("\n===== Descriptografia Completa =====")
        print(f"Texto descriptografado: {decrypted_text.decode('utf-8')}")
    except ValueError:
        print("Erro: a descriptografia falhou. Verifique as informações fornecidas.")

# Função principal que permite a seleção do usuário
def main():
    print("Selecione uma opção:")
    print("1: Criptografar um texto")
    print("2: Descriptografar um texto")

    choice = input("Digite o número da opção desejada: ")

    if choice == '1':
        print("\n=== Criptografar ===")
        measure_performance(encrypt_text)
    elif choice == '2':
        print("\n=== Descriptografar ===")
        measure_performance(decrypt_text)
    else:
        print("Escolha inválida.")

if __name__ == "__main__":
    main()
```
